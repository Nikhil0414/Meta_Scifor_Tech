```python
def add_numbers(num1, num2):
    result = num1 + num2
    return result

def subtract_numbers(num1, num2):
    result = num1 - num2
    return result

def multiply_numbers(num1, num2):
    result = num1 * num2
    return result

if __name__ == "__main__":
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    sum_result = add_numbers(num1, num2)
    print(f"The sum of {num1} and {num2} is: {sum_result}")

    difference_result = subtract_numbers(num1, num2)
    print(f"The difference between {num1} and {num2} is: {difference_result}")

    product_result = multiply_numbers(num1, num2)
    print(f"The product of {num1} and {num2} is: {product_result}")
```

```python
def calculate_simple_interest(principal, rate, time):
    simple_interest = (principal * rate * time) / 100
    return simple_interest

def calculate_compound_interest(principal, rate, time):
    compound_interest = principal * (1 + rate / 100) ** time - principal
    return compound_interest

if __name__ == "__main__":
    principal_amount = float(input("Enter the principal amount: "))
    interest_rate = float(input("Enter the interest rate: "))
    time_period = float(input("Enter the time period (in years): "))

    simple_interest_result = calculate_simple_interest(principal_amount, interest_rate, time_period)
    compound_interest_result = calculate_compound_interest(principal_amount, interest_rate, time_period)

    print(f"\nSimple Interest: {simple_interest_result}")
    print(f"Compound Interest: {compound_interest_result}")
```

```python
def toWord(number):
    num_str = str(number)
    word_list = []

    for digit in num_str:
        if digit == '0':
            word_list.append('zero')
        elif digit == '1':
            word_list.append('one')
        elif digit == '2':
            word_list.append('two')
        elif digit == '3':
            word_list.append('three')
        elif digit == '4':
            word_list.append('four')
        elif digit == '5':
            word_list.append('five')
        elif digit == '6':
            word_list.append('six')
        elif digit == '7':
            word_list.append('seven')
        elif digit == '8':
            word_list.append('eight')
        elif digit == '9':
            word_list.append('nine')

    result = '-'.join(word_list)
    return result


print(toWord(89))
print(toWord(398))
```

```python
def branch_report(branch_name, total_stock, sold_stocks):
    available_stock = total_stock - sold_stocks
    print(f"{branch_name} Branch - Available Stock: {available_stock}%")

# Example usage
branch_report("Branch A", 100, 25)
branch_report("Branch B", 80, 15)
branch_report("Branch C", 120, 30)
branch_report("Branch D", 150, 60)
```
