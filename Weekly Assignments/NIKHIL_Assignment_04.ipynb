```python
def calculate_probability(favorable_outcomes, total_outcomes):
    probability = favorable_outcomes / total_outcomes
    return round(probability, 2)

favorable_outcomes = int(input("Enter the number of favorable outcomes: "))
total_outcomes = int(input("Enter the total number of possible outcomes: "))

print(f"Probability: {calculate_probability(favorable_outcomes, total_outcomes)}")
```

```python
from datetime import date

def calculate_age(birth_date):
    today = date.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age

def check_eligibility():
    year = int(input("Enter the year of birth (YYYY): "))
    month = int(input("Enter the month of birth (MM): "))
    day = int(input("Enter the day of birth (DD): "))

    birth_date = date(year, month, day)
    age = calculate_age(birth_date)

    if age >= 18:
        print(f"Congratulations! You are eligible for a Driving License. Your age is {age} years.")
    else:
        print(f"Sorry, you are not eligible for a Driving License. Your age is {age} years.")

check_eligibility()
```

```python
class ATM:
    def __init__(self):
        self.bank_balance = 50000

    def withdraw(self, amount):
        if amount % 100 == 0 and amount <= self.bank_balance:
            self.bank_balance -= amount
            print(f"Withdrawal of {amount} successful. Remaining balance: {self.bank_balance}")
        else:
            print("Invalid amount or insufficient funds. Withdrawal failed.")

    def deposit(self, amount):
        if amount > 0:
            self.bank_balance += amount
            print(f"Deposit of {amount} successful. Updated balance: {self.bank_balance}")
        else:
            print("Invalid amount. Deposit failed.")

    def balance_enquiry(self):
        print(f"Your account balance is: {self.bank_balance}")

    def fast_cash(self, option):
        fast_cash_options = [5000, 10000, 15000, 20000, 25000, 30000, 40000, 45000, 50000]
        if option in fast_cash_options and option <= self.bank_balance:
            self.bank_balance -= option
            print(f"Fast Cash withdrawal of {option} successful. Remaining balance: {self.bank_balance}")
        else:
            print("Invalid fast cash option or insufficient funds. Fast Cash withdrawal failed.")

atm_machine = ATM()

while True:
    print("\nSelect an option:")
    print("1. Withdraw\n2. Deposit\n3. Balance Enquiry\n4. Fast Cash\n5. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        withdraw_amount = int(input("Enter the withdrawal amount: "))
        atm_machine.withdraw(withdraw_amount)
    elif choice == 2:
        deposit_amount = int(input("Enter the deposit amount: "))
        atm_machine.deposit(deposit_amount)
    elif choice == 3:
        atm_machine.balance_enquiry()
    elif choice == 4:
        fast_cash_option = int(input("Select a Fast Cash option: 5000, 10000, 15000, 20000, 25000, 30000, 40000, 45000, 50000\n"))
        atm_machine.fast_cash(fast_cash_option)
    elif choice == 5:
        print("Exiting the ATM. Thank you!")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```

```python
def calculate_simple_interest(principal, rate, time):
    return principal * rate * time / 100

def calculate_compound_interest(principal, rate, time):
    return principal * (1 + rate / 100) ** time - principal

principal_amount = 10000
yash_interest_rate = 6
vishal_interest_rate = 6
time_period = 30

yash_simple_interest = calculate_simple_interest(principal_amount, yash_interest_rate, time_period)
vishal_compound_interest = calculate_compound_interest(principal_amount, vishal_interest_rate, time_period)

difference_in_returns = vishal_compound_interest - yash_simple_interest

yash_one_year_interest = calculate_simple_interest(principal_amount, yash_interest_rate, 1)
vishal_one_year_interest = calculate_compound_interest(principal_amount, vishal_interest_rate, 1)
difference_one_year = vishal_one_year_interest - yash_one_year_interest

print(f"Yash's returns after 30 years (Simple Interest): ${yash_simple_interest:.2f}")
print(f"Vishal's returns after 30 years (Compound Interest): ${vishal_compound_interest:.2f}")
print(f"Difference in returns after 30 years: ${difference_in_returns:.2f}")
print(f"Difference in returns for one year: ${difference_one_year:.2f}")

print("\nDiscussion:")
print("The key reason for the difference in returns is that simple interest is calculated on the original principal amount each year, while compound interest is calculated on the updated principal (including the interest from previous years). This leads to compounding growth for compound interest, resulting in higher returns over time.")
```

```python
def calculate_tax(income):
    if income <= 50000:
        tax = income * 0.1
    elif 50001 <= income <= 100000:
        tax = 50000 * 0.1 + (income - 50000) * 0.15
    else:
        tax = 50000 * 0.1 + 50000 * 0.15 + (income - 100000) * 0.2
    return tax

annual_income = float(input("Enter your annual income: "))
tax_to_be_paid = calculate_tax(annual_income)
print(f"The tax to be paid based on annual income ${annual_income:.2f} is ${tax_to_be_paid:.2f}")
```

```python
def calculate_bmi(weight, height_cm):
    height_m = height_cm / 100
    bmi = weight / (height_m ** 2)
    return bmi

def interpret_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi <= 24.9:
        return "Normal weight"
    else:
        return "Overweight"

weight = float(input("Enter your weight in kilograms: "))
height_cm = float(input("Enter your height in centimeters: "))

bmi_result = calculate_bmi(weight, height_cm)
interpretation = interpret_bmi(bmi_result)

print(f"Your BMI is: {bmi_result:.2f}")
print(f"Interpretation: {interpretation}")
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Enter a year: "))

if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
