```python
import random

def guess_the_number():
    print("Welcome to the Guess the Number Game!")
    print("I have picked a random number between 1 and 100.")
    print("You have 10 chances to guess the correct number.\n")

    secret_number = random.randint(1, 100)
    attempts = 0

    while attempts < 10:
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        attempts += 1

        if guess == secret_number:
            print(f"Congratulations! You guessed the correct number {secret_number} in {attempts} attempts.")
            break
        elif guess < secret_number:
            print("Too low! Try again.")
        else:
            print("Too high! Try again.")

    else:
        print(f"\nSorry, you've run out of attempts. The correct number was {secret_number}. You lose!")

if __name__ == "__main__":
    guess_the_number()
```

```python
def adventure_game():
    print("Welcome to the Adventure Game!")

    print("\nScenario 1:")
    print("You are standing outside of your house, and you see a man running towards you, asking for urgent shelter.")
    choice1 = input("Will you provide shelter to him? (Yes / No): ").lower()

    if choice1 == "yes":
        print("\nScenario 2:")
        print("After 2 minutes, the police came to your house and ask you whether the thief was in your house or not.")
        choice2 = input("Will you say (Yes / No): ").lower()

        if choice2 == "yes":
            print("\nYou are an honest person. He was a thief. You won the Game!")
        elif choice2 == "no":
            print("\nYou denied knowing anything. The police investigate and find the thief elsewhere. You continue the adventure.")
            print("\nScenario 3:")
            print("A mysterious portal appears in your living room. Do you enter the portal? (Yes / No): ")
            choice3 = input().lower()

            if choice3 == "yes":
                print("\nScenario 4:")
                print("You find yourself in a magical land. A talking dragon approaches you. Do you try to befriend the dragon? (Yes / No): ")
                choice4 = input().lower()

                if choice4 == "yes":
                    print("\nThe dragon becomes your ally. You embark on epic adventures together. You won the Game!")
                elif choice4 == "no":
                    print("\nThe dragon gets offended and breathes fire. Game Over!")
                else:
                    print("\nInvalid choice. Game Over!")

            elif choice3 == "no":
                print("\nYou decide not to enter the portal. The portal disappears. The adventure continues.")

            else:
                print("\nInvalid choice. Game Over!")

        else:
            print("\nInvalid choice. Game Over!")

    elif choice1 == "no":
        print("\nYou chose not to provide shelter. Game Over!")
    else:
        print("\nInvalid choice. Game Over!")

adventure_game()
```

```python
import random

def choose_random_word():
    words = ['school', 'books', 'friends', 'maths', 'parents', 'zoology', 'games', 'television', 'space', 'rocket', 'chemistry', 'house', 'studies', 'project']
    return random.choice(words)

def jumble_word(word):
    jumbled_word = list(word)
    random.shuffle(jumbled_word)
    return ''.join(jumbled_word)

def play_jumbled_words_game():
    print("Welcome to the Jumbled Words Game!")

    while True:
        original_word = choose_random_word()
        jumbled = jumble_word(original_word)

        print(f"Jumbled word: {jumbled}")

        user_guess = input("Enter your guess (or type 'quit' to end the game): ")

        if user_guess.lower() == 'quit':
            print("Thanks for playing. Goodbye!")
            break

        if user_guess.lower() == original_word:
            print("Congratulations! You guessed the correct word. You get 1 point!")
        else:
            print(f"Sorry, the correct word was {original_word}. You lose this round.")

if __name__ == "__main__":
    play_jumbled_words_game()
```
