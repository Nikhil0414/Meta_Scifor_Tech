```python
def calculate_desks(rows, columns):
    total_desks = rows * columns
    return total_desks

rows = int(input("Enter the number of rows in the classroom: "))
columns = int(input("Enter the number of columns in the classroom: "))

total_desks = calculate_desks(rows, columns)

print(f"There are {total_desks} desks in the classroom.")
```

```python
def track_people(authorization_code, number_of_guests):
    if authorization_code == 1:
        print("Door Opened")
        print("Welcome to the Party! We hope you have fun.")
        number_of_guests += 1
    elif authorization_code == 0:
        print("Access Granted")
        print("Door Opened")
    else:
        print("Invalid Authorization Code")

    return number_of_guests

number_of_guests = 0

authorization_code = int(input("Enter your authorization code (0 for team members, 1 for guests): "))
number_of_guests = track_people(authorization_code, number_of_guests)

print(f"Total number of guests in the party: {number_of_guests}")
```

```python
def print_even_odd_numbers(start, end):
    print(f"Even numbers between {start} and {end}:")
    for num in range(start, end + 1):
        if num % 2 == 0:
            print(num, end=" ")

    print("\nOdd numbers between {start} and {end}:")
    for num in range(start, end + 1):
        if num % 2 != 0:
            print(num, end=" ")

start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))

print_even_odd_numbers(start_range, end_range)
```

```python
def count_words(essay):
    words = essay.split()
    return len(words)

essay_input = input("Enter the essay text: ")

word_count = count_words(essay_input)

print(f"The essay contains {word_count} words.")
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

num_to_check = int(input("Enter a number to check if it's prime: "))

if is_prime(num_to_check):
    print(f"{num_to_check} is a prime number.")
else:
    print(f"{num_to_check} is not a prime number.")
```

```python
class StudentManagement:
    def __init__(self):
        self.students = {10: "Arjun", 11: "Rajesh", 12: "Rajesh", 13: "Sathish"}
        self.current_id = 14  # Starting ID for new students

    def display_students(self):
        print("Student Details:")
        for emp_id, emp_name in self.students.items():
            print(f"{emp_id}: {emp_name}")

    def send_message(self, emp_id):
        if emp_id in self.students:
            print(f"Message sent to: {emp_id}")
        else:
            print(f"Student with ID {emp_id} not found.")

    def add_student(self, emp_name):
        new_id = self.current_id
        self.students[new_id] = emp_name
        self.current_id += 1
        print(f"Student added with ID {new_id}")

student_manager = StudentManagement()

student_manager.display_students()

student_manager.send_message(11)
student_manager.send_message(15)

student_manager.add_student("New Student 1")
student_manager.add_student("New Student 2")

student_manager.display_students()
```
