```python
pet_name = input("Enter the pet name: ")
reversed_pet_name = pet_name[::-1]
is_palindrome = pet_name.lower() == reversed_pet_name.lower()
print(is_palindrome)
```

```python
def secret_language(word):
    secret_language = ''.join([word[i + len(word)//2] + word[i] for i in range(len(word)//2)])
    return secret_language

user_input = input("Enter a word: ").split(' ')
secret_l1 = [secret_language(char) for char in user_input]
result = ' '.join(secret_l1)
print(f"Secret Language: {result}")
```

```python
def is_valid_phone_number(phone_number):
    if len(phone_number) == 10 and phone_number[0] in ['9', '8', '7'] and phone_number.isdigit():
        return True
    else:
        return False

user_input = input("Enter the phone number: ")

if is_valid_phone_number(user_input):
    print("Valid phone number!")
else:
    print("Invalid phone number. Please enter a valid 10-digit number starting with 9, 8, or 7.")
```
