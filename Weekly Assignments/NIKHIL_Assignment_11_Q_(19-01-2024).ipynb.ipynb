```python
def intersection(set1, set2):
    result = []
    for element in set1:
        if element in set2:
            result.append(element)
    return result

def complement(universal_set, set_to_complement):
    result = []
    for element in universal_set:
        if element not in set_to_complement:
            result.append(element)
    return result

set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}

intersection_result = intersection(set1, set2)
complement_result = complement(set(range(1, 11)), set1)

print("Intersection:", intersection_result)
print("Complement:", complement_result)
```

```python
def union(set1, set2):
    result = list(set1)

    for element in set2:
        if element not in set1:
            result.append(element)

    return result

A = {19, 34, 56, 7}
B = {1, 2, 5, 19, 34, 7}

union_result = union(A, B)
print("Union:", union_result)
```

```python
def sort_questions(question_list):
    n = len(question_list)
    for i in range(n):
        for j in range(0, n-i-1):
            if question_list[j] > question_list[j+1]:
                question_list[j], question_list[j+1] = question_list[j+1], question_list[j]

def find_page_number(sorted_questions, target_question):
    for i, question in enumerate(sorted_questions):
        if question == target_question:
            return i + 1
    return -1

total_questions = int(input("Enter total number of questions: "))
questions = []
print("Enter question numbers here:")
for _ in range(total_questions):
    question = int(input())
    questions.append(question)

print("Selected questions are:", questions)

sort_questions(questions)
print("Selected questions in order:", questions)

target_question = int(input("Enter the question number: "))
page_number = find_page_number(questions, target_question)
if page_number != -1:
    print(f"Solve it on page number {page_number}.")
else:
    print("You donâ€™t have to solve this question.")
```
