```python
user_input = input("Enter a list of numbers: ")
input_numbers = [int(num) for num in user_input.split()]

for i in range(len(input_numbers)):
    if input_numbers[i] % 2 != 0:
        input_numbers[i] += 1

print("Input Numbers:", input_numbers)
```

```python
user_input = input("Enter a list of numbers: ")
input_numbers = []

for char in user_input:
    if char.isdigit():
        input_numbers.append(int(char))

if len(input_numbers) >= 2:
    input_numbers[0], input_numbers[-1] = input_numbers[-1], input_numbers[0]

print("Modified List:", input_numbers)
```

```python
class HRSystem:
    def __init__(self):
        self.employee_records = []

    def display_records(self):
        print("\n Employee Records:")
        for record in self.employee_records:
            print(record)

    def correct_spelling(self, employee_id, corrected_name):
        for record in self.employee_records:
            if record.startswith(employee_id):
                self.employee_records.remove(record)
                self.employee_records.append(f"{employee_id}-{corrected_name}")
                print(f"\n Spelling corrected for {employee_id} to {corrected_name}")
                break

    def add_employee(self, employee_id, employee_name):
        self.employee_records.append(f"{employee_id}-{employee_name}")
        print(f"\n {employee_name} added to the records")

    def add_absent_employee(self, employee_id, employee_name):
        index_to_insert = int(employee_id.split('-')[0])
        self.employee_records.insert(index_to_insert - 1, f"{employee_id}-{employee_name}")
        print(f"\n {employee_name} added to the records as absent")

    def delete_employee(self, employee_id):
        for record in self.employee_records:
            if record.startswith(employee_id):
                self.employee_records.remove(record)
                print(f"\n {employee_id} deleted from the records")
                break

    def delete_all_employees(self):
        self.employee_records = []
        print("\n All employee records deleted")


# Example Usage:
hr_system = HRSystem()

# Initial records
hr_system.employee_records = ["001-Jack", "002-Jacob", "003-Crystal", "006-Kylie", "007-Chuck", "010-Andre"]

# Operations
hr_system.display_records()

# 1. Correct spelling
hr_system.correct_spelling("001", "Jackie")

# 2. Add new employees
hr_system.add_employee("011", "Emily")

# 3. Add absent employees
hr_system.add_absent_employee("005", "Tony")

# 4. Delete resigned employee
hr_system.delete_employee("007")

# Display updated records
hr_system.display_records()

# 5. Delete all records after project completion
hr_system.delete_all_employees()

# Display empty records
hr_system.display_records()
```
