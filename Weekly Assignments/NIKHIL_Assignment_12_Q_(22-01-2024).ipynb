```python
def convert_to_uppercase(input_str):
    return input_str.upper()

name = input("Enter your name: ")
dob = input("Enter your date of birth: ")
pan = input("Enter your PAN number: ")

name_upper = convert_to_uppercase(name)
dob_upper = convert_to_uppercase(dob)
pan_upper = convert_to_uppercase(pan)

print("\nInformation in Capital Letters:")
print(f"Name: {name_upper}")
print(f"Date of Birth: {dob_upper}")
print(f"PAN Number: {pan_upper}")
```

```python
def calculate_simple_interest(principal, rate, time):
    return principal * rate * time / 100

def calculate_compound_interest(principal, rate, time):
    return principal * (1 + rate / 100) ** time - principal

principal_amount = 10000
time_period = 30
simple_interest_rate = 6
compound_interest_rate = 6

simple_interest_return = calculate_simple_interest(principal_amount, simple_interest_rate, time_period)
compound_interest_return = calculate_compound_interest(principal_amount, compound_interest_rate, time_period)

difference_after_30_years = compound_interest_return - simple_interest_return
difference_in_one_year = (compound_interest_return - simple_interest_return) / time_period

print("Difference in returns after 30 years:", difference_after_30_years)
print("Difference in returns in one year:", difference_in_one_year)
```

```python
def calculate_percentage(subject_marks):
    total_marks = sum(subject_marks)
    total_subjects = len(subject_marks)
    return (total_marks / (total_subjects * 100)) * 100

subject_marks = []
for i in range(5):
    marks = float(input(f"Enter marks for subject {i + 1}: "))
    subject_marks.append(marks)

overall_percentage = calculate_percentage(subject_marks)
print("Overall Percentage:", overall_percentage)
```

```python
while True:
    n = int(input("Enter the term (n) for Fibonacci series (enter -1 to exit): "))

    if n == -1:
        break

    a, b = 0, 1
    for _ in range(n - 1):
        a, b = b, a + b

    print(f"The {n}th term of the Fibonacci series is: {a}")
```

```python
import os

def create_student_file(student_name):
    file_name = f"{student_name}_file.txt"
    with open(file_name, 'w') as file:
        file.write(f"Transaction history for {student_name}:\n")
    print(f"File '{file_name}' created successfully.")

def add_transaction(student_name, transaction):
    file_name = f"{student_name}_file.txt"
    with open(file_name, 'a') as file:
        file.write(f"{transaction}\n")
    print("Transaction added successfully.")

def read_student_file(student_name):
    file_name = f"{student_name}_file.txt"
    if os.path.exists(file_name):
        with open(file_name, 'r') as file:
            content = file.read()
        print(content)
    else:
        print(f"File for {student_name} does not exist.")

def rename_student_file(student_name, new_name):
    old_file_name = f"{student_name}_file.txt"
    new_file_name = f"{new_name}_file.txt"
    os.rename(old_file_name, new_file_name)
    print(f"File renamed to '{new_file_name}' successfully.")

def delete_student_file(student_name):
    file_name = f"{student_name}_file.txt"
    if os.path.exists(file_name):
        os.remove(file_name)
        print(f"File '{file_name}' deleted successfully.")
    else:
        print(f"File for {student_name} does not exist.")

student_name = input("Enter student's name: ")

create_student_file(student_name)
add_transaction(student_name, "Enrolled in kindergarten.")
add_transaction(student_name, "Promoted to 1st grade.")

read_student_file(student_name)

new_name = input("Enter new name for the student: ")
rename_student_file(student_name, new_name)

delete_option = input("Do you want to delete the student's file? (yes/no): ")
if delete_option.lower() == 'yes':
    delete_student_file(new_name)
```
